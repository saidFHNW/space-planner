```python
#analyzing the structure -> finding all unique top level keys

unique_keys= set()

for product in data:
    unique_keys.update(product.keys())

print(f'Unique keys: {unique_keys}')
```

```python
#counting  all categories

unique_categories = set()

for product in data:
    for category in product["categories"]:
        unique_categories.add(category["name"])

print(f'Number of unique categories: {len(unique_categories)+1}')
print(f'Unique categories: {unique_categories}')
```

```python
# SKU from category Street Workout & Fitness

sku_streetworkout_fitness = set()

for product in data:
    for category in product["categories"]:
        if category["id"] == 15:
            sku_streetworkout_fitness.add(product["sku"])

print(sku_streetworkout_fitness)
```

```python
# SKU from category Seating & Planters

sku_seating_planters = set()

for product in data:
    for category in product["categories"]:
        if category["id"] == 17:
            sku_seating_planters.add(product["sku"])

print(sku_seating_planters)
```

```python
# SKU from category Climbing & Boulder

sku_climbing_boulder = set()

for product in data:
    for category in product["categories"]:
        if category["id"] == 18:
            sku_climbing_boulder.add(product["sku"])

print(sku_climbing_boulder)
```

```python
#SKU from category "Other"

sku_other = set()

for product in data:
    for category in product["categories"]:
        if category["id"] == 16:
            sku_other.add(product["sku"])

print(sku_other)
```

```python
#removing SKU which are not worth keeping from organizedModels folder

#1. step: adding all sku in one folder

sku_all = sku_other |sku_seating_planters | sku_streetworkout_fitness | sku_climbing_boulder

print(f'number of skus: {len(sku_all)+1}')
print(f'All skus: {sku_all}')
```

```python
#2. step: finding  the files in the folder
from pathlib import Path

# Folder where the GLB files are stored
model_folder = Path(r"C:\Users\saidf\space-planner\static\organizedModels")

for sku in sku_all:
    file_path = model_folder / f"{sku}.glb"

    if file_path.exists():
        print(f"Would delete: {file_path}")
    else:
        print(f"Not found: {file_path}")
```

```python
#3.step: removing the files
for sku in sku_all:
    file_path = model_folder / f"{sku}.glb"

    if file_path.exists():
        file_path.unlink()
        print(f"Deleted: {file_path}")
    else:
        print(f"Not found: {file_path}")
```
